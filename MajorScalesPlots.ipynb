```python
scaleDict = {}

alpha = list('ABCDEFG')
scale = alpha[2:]+alpha[:2]
# start at C 
# count seven letters in mod7
# move to the V in the scale
# add sharp to last one

# naturals + C# + F#
i = 5
for k in range(8):
    i = (i+4)%7
    if k > 0:
        scale[-1] = scale[-1]+'#'
    print(scale[0],scale)
    scaleDict[scale[0]] = scale
    scale = scale[4:] + scale[:4]
        
# flats
i = 3
scale = alpha[5:]+alpha[:5]
scale[3] = 'Bb'
for k in range(7):
    i = (i+3)%7
    if k > 0:
        scale[3] = scale[3]+'b'
    print(scale[0],scale)
    scaleDict[scale[0]] = scale
    scale = scale[3:] + scale[:3]
```

```python
colorDict = {'A': 'red',
            'A#': 'orangered',
            'Bb': 'orangered',
            'B': 'orange',
            'Cb': 'orange',
            'C': 'yellow',
            'B#': 'yellow',
            'C#': 'chartreuse',
            'Db': 'chartreuse',
            'D': 'green',
            'D#': 'teal',
            'Eb': 'teal',
            'E': 'blue',
            'Fb': 'blue',
            'F': 'slateblue',
            'E#': 'indigo',
            'F#': 'purple',
            'Gb': 'purple',
            'G': 'mediumvioletred',
            'G#': 'crimson',
            'Ab': 'crimson'}
```

```python
from matplotlib.patches import Rectangle
from matplotlib.collections import PatchCollection
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Full major scale chart

'''
Nats: ABCDEFG
Sharps: C# F#
Flats: Ab Bb Cb Db Fb Gb Ab Bb
'''
fig, ax = plt.subplots(1)
fig.set_size_inches((10,10))

boxes = []
vals = scaleDict.values()
I = len(vals)
J = 7
for i,scale in enumerate(scaleDict.values()):
    for j,note in enumerate(scale):
        rect = Rectangle((j/J,1-i/I-1/I), 1/J, 1/I, facecolor=colorDict[note],alpha=0.5,edgecolor='None')
        boxes.append(rect)
        ax.add_patch(rect)
        ax.text(j/J+0.42/J-0.05/J*(len(note)>1),1-i/I-0.65/I,note,fontsize=16,weight=['normal','bold'][j==0])

ax.axvline(1/J,c='k',linewidth=1)
plt.yticks([])
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.xticks([j/J+0.5/J for j in range(7)],['I','II','III','IV','V','VI','VII'],fontsize=16)
#ax.add_collection(pc)
plt.show()

fig.savefig('scales.png',format='png',dpi=600)
```

```python
# Alphabetized & pruned down to common scales

'''
Nats: ABCDEFG
Sharps: C# F#
Flats: Ab Bb Cb Db Fb Gb Ab Bb
'''
fig, ax = plt.subplots(1)
fig.set_size_inches((10,10))

ones = ['A','Bb','B','C','C#','D','Eb','F','F#','G','Ab']

boxes = []
vals = scaleDict.values()
I = len(ones)
J = 7
for i,note in enumerate(ones):
    scale = scaleDict[note]
    for j,note in enumerate(scale):
        rect = Rectangle((j/J,1-i/I-1/I), 1/J, 1/I, facecolor=colorDict[note],alpha=0.5,edgecolor='None')
        boxes.append(rect)
        ax.add_patch(rect)
        ax.text(j/J+0.42/J-0.05/J*(len(note)>1),1-i/I-0.65/I,note,fontsize=16,weight=['normal','bold'][j==0])

ax.axvline(1/J,c='k',linewidth=1)
plt.yticks([])
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.xticks([j/J+0.5/J for j in range(7)],['I','II','III','IV','V','VI','VII'],fontsize=16)
#ax.add_collection(pc)
plt.show()

fig.savefig('scales_alpha.png',format='png',dpi=600)
```

```python
'''
Nats: ABCDEFG
Sharps: C# F#
Flats: Ab Bb Cb Db Fb Gb Ab Bb
'''
fig, ax = plt.subplots(1)
fig.set_size_inches((2,10))

inverseColorDict = {}
for k in colorDict.keys():
    c = colorDict[k]
    if c not in inverseColorDict.keys():
        inverseColorDict[c] = []
    inverseColorDict[c].append(k)

I = len(inverseColorDict.keys())
for i,c in enumerate(inverseColorDict.keys()):
        notes = ' / '.join(inverseColorDict[c]) 
        rect = Rectangle((0,1-i/I-1/I), 1, 1/I, facecolor=c,alpha=0.5,edgecolor='None')
        boxes.append(rect)
        ax.add_patch(rect)
        ax.text(0.45-0.035*(len(notes)-1),1-i/I-0.65/I,notes,fontsize=16)
plt.yticks([])
plt.xticks([])
plt.show()

fig.savefig('colorbar.png',format='png',dpi=600)
```
